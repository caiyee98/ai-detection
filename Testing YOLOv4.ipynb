```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

# display the resulting image 
def imShow(path):
    image = cv2.imread(path)
    height, width = image.shape[:2]
    resized_image = cv2.resize(image,(3*width, 3*height), interpolation = cv2.INTER_CUBIC)
    fig = plt.gcf()
    fig.set_size_inches(18, 10)
    plt.axis("off")
    plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
# Testing a new image with the trained model weights and displayed the results
!./darknet detector test /home/oppstaradmin/detection/obj.data /home/oppstaradmin/detection/yolov4_backup/model_yolov4.cfg /home/oppstaradmin/detection/yolov4_backup/model_yolov4_last.weights  /home/oppstaradmin/dataset_labelledimage/12962.jpg
imShow('predictions.jpg')
```
